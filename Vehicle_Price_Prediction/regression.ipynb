```python
data['interior_color'] = data['interior_color'].str.title()
data['interior_color'] = data['interior_color'].str.replace(r'\s+', ' ', regex=True).str.strip()
data['interior_color'] = data['interior_color'].str.replace(r'[\xa0]', '', regex=True)
```

```python
def categorize_interior_color(color):
    if pd.isna(color):
        return None  # Keeping NaN as NaN
    color = str(color).lower()
    
    if any(kw in color for kw in ['black', 'ebony', 'onyx', 'jet']):
        return 'Black'
    elif any(kw in color for kw in ['gray', 'graphite', 'pebble', 'slate', 'charcoal', 'ash']):
        return 'Gray'
    elif any(kw in color for kw in ['beige', 'tan', 'wicker', 'sandstone', 'macchiato', 'pearl']):
        return 'Beige'
    elif any(kw in color for kw in ['brown', 'walnut', 'espresso', 'java', 'cognac', 'caramel']):
        return 'Brown'
    elif any(kw in color for kw in ['red', 'ruby', 'monaco', 'burgundy']):
        return 'Red'
    elif any(kw in color for kw in ['white', 'light platinum']):
        return 'White'
    elif any(kw in color for kw in ['green', 'olive']):
        return 'Green'
    elif any(kw in color for kw in ['blue', 'navy', 'pier', 'santorini']):
        return 'Blue'
    else:
        return 'Others'
data['interior_color']=data['interior_color'].apply(categorize_interior_color)
```

```python
def clean_transmission(value):
    if pd.isna(value):
        return np.nan  
    value = value.lower().strip()  
    

    replacements = {
        'a/t': 'automatic',
        'spd': 'speed',
        '(cvt) cont var.': 'cvt',
        'cvt with xtronic': 'cvt',
        'automatic cvt': 'cvt',
        'automatic w/paddle shifters': 'automatic',
        '9 spd automatic': '9-speed automatic',
        '6-spd aisin f21-250 phev auto trans': '6-speed automatic',
        '7-speed dsga? automatic w/ 4mo': '7-speed dsg automatic',
        '62 kwh battery': 'electric'
    }
    

    for key, val in replacements.items():
        value = value.replace(key, val)
    

    return value.title()

data['transmission'] = data['transmission'].apply(clean_transmission)
```

# Splitting the data

```python
from sklearn.model_selection import train_test_split


x = data.drop(columns=['price_boxcox'])  
y = data['price_boxcox']  


x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

print(x_train.shape, x_test.shape) 
```

# Scaling the data

```python
from sklearn.preprocessing import StandardScaler
num_cols = x_train.select_dtypes(include=['float64', 'int64']).columns
scaler = StandardScaler()
x_train[num_cols] = scaler.fit_transform(x_train[num_cols])
x_test[num_cols] = scaler.transform(x_test[num_cols])
```

# Feature Selection & Importance

```python
from sklearn.linear_model import LassoCV


lasso = LassoCV(cv=5).fit(x_train, y_train)
important_features = np.where(lasso.coef_ != 0)[0]


x_train_new = x_train.iloc[:, important_features]
x_test_new = x_test.iloc[:, important_features]
```

```python
from xgboost import XGBRegressor

xgb = XGBRegressor()
xgb.fit(x_train_new, y_train)
```

```python
plt.figure(figsize=(20,15))
plt.barh(x_train_new.columns, xgb.feature_importances_)
plt.show()
```

```python
feature_importance_df = pd.DataFrame({
    'Feature': xgb.feature_names_in_,
    'Importance': xgb.feature_importances_
})

feature_importance_df = feature_importance_df.sort_values(by="Importance", ascending=False)

feature_importance_df
```

```python
threshold = 0.01  
selected_features = feature_importance_df[feature_importance_df['Importance'] > threshold]['Feature'].tolist()
selected_features
```

```python
x_train_selected = x_train_new[selected_features]
x_test_selected = x_test_new[selected_features]
```

# Model Traning and Evaluation

```python
from sklearn.model_selection import GridSearchCV, cross_val_score
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from catboost import CatBoostRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
results = {}


def evaluate_model(model, x_train, y_train, x_test, y_test, name):
    y_pred = model.predict(x_test)
    mae = mean_absolute_error(y_test, y_pred)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    r2 = r2_score(y_test, y_pred)
    results[name] = {'MAE': mae, 'RMSE': rmse, 'R2': r2}
    print(f"{name} - MAE: {mae:.2f}, RMSE: {rmse:.2f}, R2: {r2:.4f}")
```

```python
# **Linear Regression**
lr = LinearRegression()
lr.fit(x_train_selected, y_train)
evaluate_model(lr, x_train_selected, y_train, x_test_selected, y_test, "Linear Regression")
```

```python
# **Ridge Regression**
ridge_params = {'alpha': [0.1, 1, 10, 100]}
ridge = GridSearchCV(Ridge(), ridge_params, cv=5)
ridge.fit(x_train_selected, y_train)
evaluate_model(ridge.best_estimator_, x_train_selected, y_train,x_test_selected, y_test, "Ridge Regression")
```

```python
# **Lasso Regression (L1 Regularization)**
lasso_params = {'alpha': [0.1, 1, 10, 100]}
lasso = GridSearchCV(Lasso(), lasso_params, cv=5)
lasso.fit(x_train_selected, y_train)
evaluate_model(lasso.best_estimator_, x_train_selected, y_train, x_test_selected, y_test, "Lasso Regression")
```

```python
# **Support Vector Regression (SVR)**
svr_params = {'C': [0.1, 1, 10], 'epsilon': [0.01, 0.1, 0.2], 'kernel': ['linear', 'rbf']}
svr = GridSearchCV(SVR(), svr_params, cv=5)
svr.fit(x_train_selected, y_train)
evaluate_model(svr.best_estimator_, x_train_selected, y_train, x_test_selected, y_test, "SVR")
```

```python
# **Decision Tree Regressor**
dt_params = {'max_depth': [3, 5, 10, None], 'min_samples_split': [2, 5, 10]}
dt = GridSearchCV(DecisionTreeRegressor(), dt_params, cv=5)
dt.fit(x_train, y_train)    
evaluate_model(dt.best_estimator_, x_train, y_train, x_test, y_test, "Decision Tree")
```

```python
# **XGBoost Regressor**
xgb_params = {'n_estimators': [100, 200], 'max_depth': [3, 5, 10], 'learning_rate': [0.01, 0.1, 0.2]}
xgb = GridSearchCV(XGBRegressor(objective='reg:squarederror'), xgb_params, cv=5)
xgb.fit(x_train, y_train)
evaluate_model(xgb.best_estimator_, x_train, y_train, x_test, y_test, "XGBoost")
```

```python
# **CatBoost Regressor**
catboost = CatBoostRegressor(verbose=0)
catboost.fit(x_train, y_train)
evaluate_model(catboost, x_train, y_train, x_test, y_test, "CatBoost")
```
